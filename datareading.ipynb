```python
import pandas as pd

df = pd.read_csv("./data/daily_sales_data_0.csv")
df.head()
```

```python
df = df[df['product'].str.contains('pink morsel')==True]
```

```python
df.head()
```

```python
df['price'] = df['price'].str.replace('$', '')

df.head()
```

```python
df['price'] = df['price'].astype(float)
```

```python
df['sales'] = df['price'] * df['quantity']
```

```python
df.info()
```

```python
cols = ['sales','date','region']

df = df[cols]
```

```python
df.head()
```

```python
df.to_csv(r'./data/df1.csv', index = False)
```
